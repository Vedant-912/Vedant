```python
# prompt: print hello world using rot13

def rot13(message):
  alphabets = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
  rot_alphabets = alphabets[13:] + alphabets[:13]
  trans_table = str.maketrans(alphabets, rot_alphabets)
  return message.translate(trans_table)
print(rot13("Hello World"))
```
